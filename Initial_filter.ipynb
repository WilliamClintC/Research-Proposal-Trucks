```python
import pandas as pd

file_path = r'C:\Users\clint\Desktop\Research Proposal Trucks\Crash_File_20241222.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df
```

```python
df = df[df['VEHICLE_CONFIGURATION_ID'] >= 3]
df
```

```python
df.columns
```

```python
unique_entries = df['CRASH_EVENT_SEQ_ID_DESC'].unique()
unique_entries
```

```python
df_filtered = df[df['CRASH_EVENT_SEQ_ID_DESC'].str.contains('14', na=False)]
df_filtered
```

```python
columns_to_drop = ['CHANGE_DATE', 'FINAL_STATUS_DATE', 'TRAFFICWAY_ID', 'ACCESS_CONTROL_ID', 'ROAD_SURFACE_CONDITION_ID', 'VEHICLE_LIC_STATE', 'VEHICLE_HAZMAT_PLACARD', 'WEATHER_CONDITION_ID', 'LIGHT_CONDITION_ID', 'HAZMAT_RELEASED', 'AGENCY', 'TOW_AWAY', 'FEDERAL_RECORDABLE', 'STATE_RECORDABLE', 'SNET_VERSION_NUMBER', 'SNET_SEQUENCE_ID', 'TRANSACTION_DATE']

df_filtered = df_filtered.drop(columns=columns_to_drop)
df_filtered
```

```python
additional_columns_to_drop = ['UPLOAD_FIRST_BYTE', 'UPLOAD_DOT_NUMBER', 'UPLOAD_SEARCH_INDICATOR', 'UPLOAD_DATE', 'ADD_DATE', 'CRASH_CARRIER_ID', 'CRASH_CARRIER_NAME', 'CRASH_CARRIER_STREET', 'CRASH_CARRIER_CITY', 'CRASH_CARRIER_CITY_CODE', 'CRASH_CARRIER_STATE', 'CRASH_CARRIER_ZIP_CODE', 'CRASH_COLONIA', 'DOCKET_NUMBER', 'CRASH_CARRIER_INTERSTATE', 'NO_ID_FLAG']

df_filtered = df_filtered.drop(columns=additional_columns_to_drop)
df_filtered
```

```python
df_filtered.columns.tolist()
```

```python
columns_to_remove = ['REPORT_NUMBER', 'REPORT_TIME', 'DOT_NUMBER', 'TRUCK_BUS_IND', 'GVW_RATING_ID','STATE_NUMBER',	'STATE_ISSUING_NUMBER']
df_filtered = df_filtered.drop(columns=columns_to_remove)
df_filtered
```
