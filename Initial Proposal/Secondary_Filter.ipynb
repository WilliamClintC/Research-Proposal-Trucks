```python
import pandas as pd

# Load the CSV file
file_path = r'C:\Users\clint\Desktop\Research-Proposal-Trucks\filtered_data.csv'
filtered_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
filtered_data
```

```python
filtered_data_sorted = filtered_data.sort_values(by='REPORT_DATE')
filtered_data_sorted = filtered_data_sorted.drop(columns=['CI_STATUS_CODE', 'TRANSACTION_CODE','CARGO_BODY_TYPE_ID','VEHICLE_IDENTIFICATION_NUMBER','VEHICLE_LICENSE_NUMBER','VEHICLE_CONFIGURATION_ID'])
filtered_data_sorted
```

```python
import pandas as pd

def process_dataframe(df):
    # Sort the dataframe by "CRASH_ID"
    df = df.sort_values(by="CRASH_ID", ascending=False).reset_index(drop=True)
    
    # Initialize a list to keep the rows to retain
    rows_to_keep = []
    
    # Iterate over the dataframe
    i = 0
    while i < len(df):
        current_row = df.iloc[i]
        keep_row = True
        
        # Check the next 5 rows
        for j in range(1, 6):
            if i + j < len(df):
                next_row = df.iloc[i + j]
                if (current_row["REPORT_STATE"] == next_row["REPORT_STATE"] and
                    current_row["REPORT_DATE"] == next_row["REPORT_DATE"] and
                    current_row["CITY"] == next_row["CITY"] and
                    current_row["COUNTY_CODE"] == next_row["COUNTY_CODE"]):
                    keep_row = False
                    break
        
        # If no match found in the next 5 rows, keep the current row
        if keep_row:
            rows_to_keep.append(current_row)
        
        i += 1
    
    # Create a new dataframe with the rows to keep
    result_df = pd.DataFrame(rows_to_keep)
    
    return result_df


result_df = process_dataframe(filtered_data_sorted)
result_df
```

```python
result_df.to_csv('result_df.csv', index=False)
```

```python
import pandas as pd

def find_matching_entries(df):

    # Step 2: Sort by "CRASH_ID" column
    df_sorted = df.sort_values(by='CRASH_ID')

    # Step 3-5: Check for matching entries in the specified columns
    matching_entries = []

    for i in range(len(df_sorted)):
        current_row = df_sorted.iloc[i]
        for j in range(1, 6):
            if i + j < len(df_sorted):
                next_row = df_sorted.iloc[i + j]
                if (current_row['REPORT_STATE'] == next_row['REPORT_STATE'] and
                    current_row['REPORT_DATE'] == next_row['REPORT_DATE'] and
                    current_row['CITY'] == next_row['CITY'] and
                    current_row['COUNTY_CODE'] == next_row['COUNTY_CODE']):
                    matching_entries.append(current_row)
                    matching_entries.append(next_row)

    # Remove duplicates from matching_entries
    matching_entries_df = pd.DataFrame(matching_entries).drop_duplicates()

    return matching_entries_df

matching_entries_df = find_matching_entries(result_df)
matching_entries_df
```
